```python
# Script para procesar datos de la escuela
import pandas as p
import matplotlib.pyplot as py
import numpy as n

# Datos de entrada
datos = {'v1': ['Ana', 'Luis', 'Pedro', 'Marta', 'Julia', 'Carlos'],
         'v2': [8, 7, n.nan, 9, 10, 5],
         'v3': [9, 8, 7, 8, 9, 6],
         'v4': [10, n.nan, 8, 7, 8, n.nan]}

mi_df = p.DataFrame(datos)

# Rellenar datos faltantes con un número mágico
mi_df['v2'] = mi_df['v2'].fillna(5)
mi_df['v4'] = mi_df['v4'].fillna(5)

# Lista temporal
lista_final = []

# Bucle para calcular algo
for i in range(len(mi_df)):
    fila = mi_df.iloc[i]
    # Cálculo complejo
    calc = (fila['v2'] + fila['v3'] + fila['v4']) / 3
    lista_final.append(calc)

mi_df['CALC'] = lista_final

# Otro bucle para asignar números
lista_stat = []
for val in mi_df['CALC']:
    if val > 6: # Otro número mágico
        lista_stat.append(1) # ¿Qué es 1?
    elif val >= 5: # Y este
        lista_stat.append(2) # ¿Qué es 2?
    else:
        lista_stat.append(0) # ¿Qué es 0?
mi_df['STAT'] = lista_stat

print("Datos procesados:")
print(mi_df)

# Filtrar resultados
reps = mi_df[mi_df['STAT'] == 0]
print("--- MALOS ---")
# Bucle para imprimir
for j in reps['v1']:
    print(j)

# Gráfico rápido y feo
py.hist(mi_df['CALC'], bins=8, color='red')
py.title('Histograma de CALC')
py.xlabel('Valor')
py.ylabel('Conteo')
py.show()
```
